```python
import pandas as pd
import numpy as np
from bokeh.client import push_session
from bokeh.charts import output_notebook, output_file, show, Bar, Scatter, Histogram, TimeSeries
from bokeh.layouts import widgetbox
from bokeh.models.widgets import Dropdown
from bokeh.plotting import figure
from bokeh.models import Range1d, HoverTool, ResizeTool
output_notebook()
```

```python
 BBobjectsDF= pd.read_csv('../Data/BBobjects.csv')
```

```python
BBobjectsDF
```

```python
def testRun(propTyp):
    ProperTable = BBobjectsDF.filter(regex=propTyp, axis=0)
    return ProperTable
```

```python
testRun("MI_O")
```

```python
output_file("dropdown.html")

menu = [("Requirement","MI_Requirement"),("Objective", "MI_Objective"), ("Operation", "MI_Operation"),("Platform", "MI_Platform")]
dropdown = Dropdown(label="Property Type", button_type="warning", menu=menu)

show(widgetbox(dropdown))
dropdown.on_click(testRun(str(dropdown)))
```
