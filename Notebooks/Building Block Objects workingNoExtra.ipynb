#### The purpose of this notebook is to read in a csv and create a function that is driven by dropdown selection. pandas Numpy and Bokeh should do the trick...

```python
import pandas as pd
import numpy as np
from ipykernel import kernelapp as app
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from ipywidgets import Button, Layout
```

### Reading in the CSV containing the building block objects.

```python
 BBobjectsDF= pd.read_csv('../Data/BBobjects.csv')
```

### Here is the Dataframe it has four columns, PropertyType, ClassName, Element, and Role.

```python
BBobjectsDF
```

### This function looks for rows that contain a specific PropertyType. It functions by looking for a string that contains a given substring which is the only argument of the function.

```python
def PropertyTypeTable(PropertyType):
    ProperTable = BBobjectsDF[BBobjectsDF.PropertyType.str.contains( PropertyType )]
    return ProperTable
```

### This input creates a dropdown the user can select their desired PropertyType from

```python
interact(PropertyTypeTable, PropertyType=["Requirement", "Objective", "Operation", "Platform"])
```
